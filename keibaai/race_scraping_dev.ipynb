```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import urllib
from bs4 import BeautifulSoup
```

```python
base_url = "https://race.netkeiba.com/race/shutuba.html"
payload = {"race_id": "202110010611"}
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)

qs = urllib.parse.urlencode(payload)
driver.get(base_url + "?" + qs)
```

```python
html = driver.page_source
html
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
# RaceData
race_data = soup.select_one(".RaceData01")
race_data.contents
```

```python
import re
for c in race_data.contents:
    m = re.search(r"\d+:\d+", str(c))
    if m:
        print(m.group())
    m = re.search(r"天候:([^ \t/<]+)", str(c))
    if m:
        print(m.group(1))
    m = re.search(r"馬場:([^ \t/<]+)", str(c))
    if m:
        print(m.group(1))
```

```python
# RaceData
race_data = soup.select_one(".RaceData02")
race_data.contents
```

```python
for c in race_data.contents:
    m = re.search(r"(\d+)回", str(c))
    if m:
        print(m.group(1))
    m = re.search(r"(\d+)日目", str(c))
    if m:
        print(m.group(1))
    m = re.search(r"[^ \t<>/]+歳[^ \t<>/]*", str(c))
    if m:
        print(m.group())
    m = re.search(r"(\d+)頭", str(c))
    if m:
        print(m.group(1))
    m = re.search(r"馬場:([^ \t/<]+)", str(c))
    if m:
        print(m.group(1))
```
